# Steps
- [X] Load a CSV file
- [X] Create a cosine similarity function wich as input ask for 2 dataframes
- [X] Create a matrix of all the cosine similarity per each element 
- [X] Draw the heatmap with that information

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import the dataset
dataset = pd.read_csv('database.csv', index_col=0)
```

```python
# Cosine similarity function
def cosine_similarity (_a, _b):
    return np.dot(_a, _b) / (np.linalg.norm(_a)*np.linalg.norm(_b))
```

```python
# Create a matrix based on the result of the cosine similarity within every value
mat = []
for i in dataset.index:
    row = []
    for j in dataset.index:
        _a = dataset.loc[i].array
        _b = dataset.loc[j].array
        row.append(cosine_similarity(_a, _b))
    mat.append(row)
```

```python
# Getting all the index (the names) as a list
index_list = dataset.index.values.tolist()
```

```python
# Convert the matrix into a Data Frame
matrix = pd.DataFrame(mat, index_list, index_list)
```

```python
# Create the plot
plt.xticks(range(len(index_list)),index_list,rotation=90)
plt.yticks(range(len(index_list)),index_list)
plt.imshow(matrix, cmap='hot',interpolation="nearest")
```
